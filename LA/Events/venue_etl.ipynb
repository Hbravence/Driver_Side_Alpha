```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

url = "https://api.seatgeek.com/2/venues"
#get api Keys
client_id = os.getenv("client_id_venues")
# this pulls venues for the Kansas City Metro

def get_LA_venues(client_id, url):
    results = []
    page = 1
    per_page = 100  # Adjust based on API limits

    while True:
        params = {
            "city": "Los Angeles",
            "state": "CA",
            "client_id": client_id,
            "page": page,
            "per_page": per_page
        }

        res = requests.get(url, params=params)

        if res.status_code != 200:
            print(f"Error: {res.status_code} - {res.text}")
            break

        data = res.json()
        venues = data.get("venues", [])
        results.extend(venues)

        print(f"Fetched page {page} with {len(venues)} venues")

        # Stop if fewer results than per_page (last page)
        if len(venues) < per_page:
            break

        page += 1

    return results



LA_venues = get_LA_venues(client_id,url)
```

```python
import pandas as pd
#JSON is then Normalized into flattened data file 
LA_venues_df = pd.json_normalize(LA_venues)

LA_venues_df.head()
```

```python
LA_venues_df.info()
LA_venues_df = LA_venues_df[LA_venues_df['has_upcoming_events'] == True]
```

```python
#Venue is then saved to a file folder so it can be called for the event's scripts. 
#this file wont be updated frequently, due to the nature of venues not changing frequently
LA_venues_df.to_csv("events_data/venues_data/LA_Venues.csv",index=False)
```
